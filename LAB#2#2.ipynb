```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cancer = pd.read_csv('data.csv')
```

```python
cancer.head()
```

```python
cancer.isnull().sum()
```

```python
# Removing the last column with all null
cancer = cancer.dropna(axis = 1)
```

```python
cancer.iloc[:,1:32].describe()
```

```python
df_corr = cancer.iloc[:,1:32].corr()
```

```python
# we can see that correlation b/w many fieds > 0.9..causing multi collinearity..
sns.heatmap(df_corr.corr())
```

```python
#collecting the field names having > 0.9 correlation coefficient
correlated_features = set()
for i in range(len(df_corr.columns)):
    for j in range(i):
        if (df_corr.iloc[i, j]) > 0.90:
            colname = df_corr.columns[i]
            correlated_features.add(colname)
```

```python
len(correlated_features)
```

```python
print(correlated_features)
```

```python
cancer.drop(labels=correlated_features, axis=1, inplace=True)
```

```python
# Resultant 22 columns after removing 10 columns..
cancer
```

```python
from sklearn.model_selection import train_test_split
```

```python
cancer.iloc[:,1]
```

```python
y = cancer.iloc[:,1]
X = cancer.iloc[:,2:]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.001,0.0001]}
```

```python
grid = GridSearchCV(SVC(),param_grid,verbose = 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
